```python
import seaborn as sns
import numpy as np
```

```python
sns.get_dataset_names()
```

```python
titanic = sns.load_dataset("titanic")
titanic
iris = sns.load_dataset("iris")
iris
```

```python
sns.scatterplot(x="survived",y="age",data=titanic,hue="who",size="class",palette="YlGnBu")
```

```python
sns.histplot(titanic['age'],kde=True,bins=15)
```

```python
sns.barplot(x="who",y="survived",data=titanic,palette="YlGnBu")
```

```python
sns.boxplot(x="survived",y="pclass",data=titanic,hue="who",palette="YlGnBu")
```

```python
sns.stripplot(x="survived",y="pclass",data=titanic,hue="who",palette="YlGnBu",dodge=True)
```

```python
sns.jointplot(x="survived",y="pclass",data=titanic,cmap="YlGnBu")#kind="reg" it is depend on the data of the table that data in integer form only
```

```python
sns.pairplot(titanic.select_dtypes(['number']),hue="pclass")
```

```python
numeric_titanic = titanic.select_dtypes(include=[np.number])
numeric_titanic.corr()
```

```python
numeric_titanic = titanic.select_dtypes(include=[np.number])
sns.heatmap(numeric_titanic.corr(), annot=True, cmap="icefire")
```
